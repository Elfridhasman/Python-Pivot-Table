```python
import pandas as pd
```

```python
dataset = 'sample_pivot.xlsx'
df = pd.read_excel(dataset)
```

```python
df.head()
```

```python
pivot1 = pd.pivot_table(df, index='Region')
pivot1
```

```python
pivot2 = pd.pivot_table(df, index='Region', values='Sales')
pivot2
```

```python
pivot3 = pd.pivot_table(
    df, index='Region', values='Sales', aggfunc=['sum', 'mean'])
pivot3
```

```python
pivot4 = pd.pivot_table(df, index='Region',
                        columns='Type', 
                        values='Sales')
pivot4
```

```python
pivot5 = pd.pivot_table(df, index='Region',
                        columns='Type',
                        values='Units')
pivot5
```

```python
pivot6 = pd.pivot_table(df, index='Region',
                        columns='Type',
                        values='Units',
                        fill_value=0)
pivot6
```

```python
pivot7 = pd.pivot_table(df, index='Region',
                        columns='Type',
                        values='Units',
                        fill_value=0,
                        margins=True)
pivot7
```

```python
pivot8 = pd.pivot_table(df, index='Region',
                        columns='Type',
                        values='Sales',
                        aggfunc='sum',
                        margins_name='Total Sales')
pivot8
```
